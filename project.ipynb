```python
import pandas as pd
df_9192 = pd.read_csv('michaelJordanStats9192.csv')
```

```python
df_9192.head()
```

```python
df_9192.rename(columns={'Unnamed: 5': 'Home/Away'}, inplace=True)
df_9192.rename(columns={'Unnamed: 7': 'Spread'}, inplace=True)
df_9192["Home/Away"].replace({"@": "Away"}, inplace=True)
df_9192["Home/Away"].fillna("Home", inplace=True)
df_9192.head()
```

```python
df_9192['W/L'] = df_9192['Spread'].str.extract('([W|L])', expand=True)
df_9192.head()
```

```python
training_columns = ['Home/Away', 'PTS', 'AST', 'STL', 'BLK', 'TOV', 'FG%']
x = df_9192[training_columns]
y = df_9192['W/L']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x['Home/Away'] = le.fit_transform(x['Home/Away'])
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=1)
rfc.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, rfc.predict(x_test))
```
