```python
import pandas as pd
Giannis_2021_2022 = pd.read_csv('Giannis2021-22.csv')
Giannis_2022_2023 = pd.read_csv('Giannis2022-23.csv')
```

```python
Giannis_2021_2022.head()
```

```python
Giannis_2022_2023.head()
```

```python
Giannis_2021_2022.rename(columns={'Unnamed: 5': 'Home/Away'}, inplace=True)
Giannis_2021_2022.rename(columns={'Unnamed: 7': 'Spread'}, inplace=True)
Giannis_2021_2022["Home/Away"].replace({"@": "Away"}, inplace=True)
Giannis_2021_2022["Home/Away"].fillna("Home", inplace=True)
Giannis_2021_2022.head()
```

```python
Giannis_2022_2023.rename(columns={'Unnamed: 5': 'Home/Away'}, inplace=True)
Giannis_2022_2023.rename(columns={'Unnamed: 7': 'Spread'}, inplace=True)
Giannis_2022_2023["Home/Away"].replace({"@": "Away"}, inplace=True)
Giannis_2022_2023["Home/Away"].fillna("Home", inplace=True)
Giannis_2022_2023.head()
```

```python
Giannis_2021_2022['W/L'] = Giannis_2021_2022['Spread'].str.extract('([W|L])', expand=True)
Giannis_2022_2023['W/L'] = Giannis_2022_2023['Spread'].str.extract('([W|L])', expand=True)
Giannis_2021_2022.head()
```

```python
# data_features = list(Giannis_2021_2022.columns)
# data_features.remove('W/L')
# data_features.remove('Spread')
```

```python
data_features = ['Home/Away', 'Opp', 'MP', 'FG', 'FG%', '3P', 'FT', 'FTA', 'TRB', 'AST', 'PTS', 'GmSc']
Giannis_2021_2022_x = Giannis_2021_2022[data_features]
Giannis_2021_2022_y = Giannis_2021_2022['W/L']
Giannis_2022_2023_x = Giannis_2022_2023[data_features]
Giannis_2022_2023_y = Giannis_2022_2023['W/L']
```

```python
import itertools
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
Giannis_2021_2022_x['Home/Away'] = le.fit_transform(Giannis_2021_2022_x['Home/Away'])
Giannis_2021_2022_x['Opp'] = le.fit_transform(Giannis_2021_2022_x['Opp'])
Giannis_2021_2022_x['MP'] = le.fit_transform(Giannis_2021_2022_x['MP'])
Giannis_2021_2022_y = le.fit_transform(Giannis_2021_2022_y)
Giannis_2022_2023_x['Home/Away'] = le.fit_transform(Giannis_2022_2023_x['Home/Away'])
Giannis_2022_2023_x['Opp'] = le.fit_transform(Giannis_2022_2023_x['Opp'])
Giannis_2022_2023_x['MP'] = le.fit_transform(Giannis_2022_2023_x['MP'])
Giannis_2022_2023_y = le.fit_transform(Giannis_2022_2023_y)
```

```python
feature_subsets = [list(subset) for i in range(len(data_features) + 1) for subset in itertools.combinations(data_features, i)]
feature_subsets.remove([])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LogisticRegression
```

```python
rfc_mae = 1
for subset in feature_subsets:
    x_train, x_test, y_train, y_test = train_test_split(Giannis_2021_2022_x[subset], Giannis_2021_2022_y, test_size=0.2, random_state=0)
    rfc = RandomForestClassifier(random_state=1)
    rfc.fit(x_train, y_train)
    subset_mae = mean_absolute_error(y_test, rfc.predict(x_test))
    if subset_mae < rfc_mae:
        rfc_mae = subset_mae
        rfc_optimal_subset = subset
```

```python
lr_mae = 1
for subset in feature_subsets:
    x_train, x_test, y_train, y_test = train_test_split(Giannis_2021_2022_x[subset], Giannis_2021_2022_y, test_size=0.2, random_state=5)
    lr = LogisticRegression(max_iter=len(feature_subsets),random_state=1)
    lr.fit(x_train, y_train)
    subset_mae = mean_absolute_error(y_test, lr.predict(x_test))
    if subset_mae < lr_mae:
        lr_mae = subset_mae
        lr_optimal_subset = subset
```
