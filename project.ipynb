```python
import pandas as pd
df_train = pd.read_csv('Giannis2021-22.csv')
df_test = pd.read_csv('Giannis2022-23.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_train.rename(columns={'Unnamed: 5': 'Home/Away'}, inplace=True)
df_train.rename(columns={'Unnamed: 7': 'Spread'}, inplace=True)
df_train["Home/Away"].replace({"@": "Away"}, inplace=True)
df_train["Home/Away"].fillna("Home", inplace=True)
df_train.head()
```

```python
df_test.rename(columns={'Unnamed: 5': 'Home/Away'}, inplace=True)
df_test.rename(columns={'Unnamed: 7': 'Spread'}, inplace=True)
df_test["Home/Away"].replace({"@": "Away"}, inplace=True)
df_test["Home/Away"].fillna("Home", inplace=True)
df_test.head()
```

```python
df_train['W/L'] = df_train['Spread'].str.extract('([W|L])', expand=True)
df_test['W/L'] = df_test['Spread'].str.extract('([W|L])', expand=True)
df_train.head()
```

```python
data_columns = ['Home/Away', 'PTS', 'AST', 'STL', 'BLK', 'TOV', 'FG%']
x_train = df_train[data_columns]
x_test = df_test[data_columns]
y_train = df_train['W/L']
y_test = df_test['W/L']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_train['Home/Away'] = le.fit_transform(x_train['Home/Away'])
y_train = le.fit_transform(y_train)
x_test['Home/Away'] = le.fit_transform(x_test['Home/Away'])
y_test = le.fit_transform(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=1)
rfc.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, rfc.predict(x_test))
```
